```python
#now we remove the comma from the cost column
zomato['cost']=zomato['cost'].astype(str)
zomato['cost']=zomato['cost'].apply(lambda x:x.replace(',','.'))
zomato['cost']=zomato['cost'].astype(float)
zomato.info()
zomato.head(20)
```

```python
zomato['rate'].unique()
```

```python
#Removing '/5' from Rate column
zomato['rate']=zomato['rate'].astype(str)
zomato.loc[(zomato['rate']!="NEW") & (zomato['rate']!='-') ].reset_index(drop=True) # NEW is just an placeholder value
zomato['rate']=zomato['rate'].apply(lambda x:x.replace('/5',' ')).str.strip()
zomato[['rate']].head()
```

```python
#Adjust the column names
zomato['name']=zomato['name'].apply(lambda x:x.title())
zomato['online_order'].replace(('Yes','No'),(True,False),inplace=True)
zomato['book_table'].replace(('Yes','No'),(True,False),inplace=True)
zomato[['online_order','book_table']].head(5)
```

```python
zomato['cost'].unique()
```

```python
def Encode(zomato):
  for column in zomato.columns [~zomato.columns.isin(['rate','cost','votes'])]:
    zomato[column]=zomato[column].factorize()[0]
  return zomato

zomato_en=Encode(zomato.copy())
zomato_en

```

```python
#Get correlation between different variables
import matplotlib.pyplot as plt
import seaborn as sns

#errors='coerce':
#If there are any values in the cost column that cannot be converted to a numeric type (e.g., strings that dont represent numbers, or other invalid data), errors='coerce' will convert those invalid entries to NaN (Not a Number).

zomato_en['rate']=pd.to_numeric(zomato_en['rate'] ,errors='coerce')
zomato_en['cost']=pd.to_numeric(zomato_en['cost'],errors='coerce')
corr=zomato_en.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr,annot=True)
plt.savefig("correlatin.png")
plt.show()
zomato_en.columns
```

## **DATA VISUALIZATION**
###   **Table booking Rate vs Rate**

```python
plt.rcParams['figure.figsize'] = (13, 9)
Y = pd.crosstab(zomato['rate'], zomato['book_table'])
Y.div(Y.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True,color=['red','yellow'])
plt.title('table booking vs rate', fontweight = 30, fontsize = 20)
plt.legend(loc="upper right")
plt.savefig("Table_Booking_Rate.png")
plt.show()
```

### **Location and Rating**

```python
loc_plt=pd.crosstab(zomato['rate'],zomato['city'])
loc_plt.plot(kind='bar',stacked=True);
plt.title('Location - Rating',fontsize=15,fontweight='bold')
plt.ylabel('Location',fontsize=10,fontweight='bold')
plt.xlabel('Rating',fontsize=10,fontweight='bold')
plt.xticks(fontsize=10,fontweight='bold')
plt.yticks(fontsize=10,fontweight='bold');
plt.legend().remove();
plt.savefig("Location Rating.png")
```
